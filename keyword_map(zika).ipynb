```python
import numpy as np
from tqdm import tqdm
from copy import deepcopy
import math
```

```python
lookup_file = open('./LOOKUP-GKGTHEMES.txt', 'r')
lookup_gkg_themes = []
theme2id = {}
for i, line in enumerate(lookup_file.readlines()):
    lookup_theme = line.strip().split('\t')[0]
    lookup_gkg_themes.append(lookup_theme)
    theme2id[lookup_theme] = i
lookup_file.close()
```

```python
len(lookup_gkg_themes) # number of nodes
```

```python
theme2id['EPU_POLICY']
```

```python
theme_graph = np.zeros([len(lookup_gkg_themes), len(lookup_gkg_themes)])
theme_graph2 = deepcopy(theme_graph)
```

```python
themes_not_in_lookup = []

file = open('gkg_dataset/20150907.gkg.csv', 'r')
for i, line in tqdm(enumerate(file.readlines())):
    if i == 0:
        continue
    news_themes = line.strip().split('\t')[3].rstrip(';').split(';')
    for theme in news_themes:
        if theme and theme not in lookup_gkg_themes:
            if theme not in themes_not_in_lookup:
                themes_not_in_lookup.append(theme)
        elif theme:
            for theme_ in news_themes:
                if not theme_ or theme_ not in lookup_gkg_themes:
                    continue
                if theme != theme_:
                    theme_graph[theme2id[theme], theme2id[theme_]] += 1
```

```python
# average number of edges (per node)
theme_graph.sum()/len(lookup_gkg_themes)/2
```

```python
before_graph = theme_graph
before_sum = before_graph.sum(axis = 0)
```

```python
topk_indices = theme_graph.sum(axis=1).argsort()[-100:][::-1]
# print(topk_indices)
for key, value in theme2id.items():
    if value in topk_indices:
        print(key)
```

```python
after_graph = theme_graph2
after_sum = after_graph.sum(axis = 0)
after_sum
```

```python
ratio = after_sum/before_sum
```

```python
up_list = []
for them, i in zip(lookup_gkg_themes, ratio):
    if not math.isnan(i): #nan   
        up_list.append((i, them))  
up_list.sort(reverse = True)
for i in up_list:
    print(i)
```

## normalizing the matrix(with the number of edges)

```python
n_edges = theme_graph.sum()/2
norm1 = theme_graph / n_edges
norm1_sum = norm1.sum(axis = 0)

n_edges = theme_graph2.sum()/2
norm2 = theme_graph2 / n_edges
norm2_sum = norm2.sum(axis = 1)

ratio = norm2_sum / norm1_sum

up_list = []
for them, i in zip(lookup_gkg_themes, ratio):
    if not math.isnan(i): #nan   
        up_list.append((i, them)) 
        
up_list.sort(reverse = True)

n = 0
for i in up_list:
    if i[0] > 100000000:
        continue
    else:
        if n < 30:
            print(i)
            n += 1
```

## normalizing the matrix(with Max degree)

```python
max_degree1 = np.max(theme_graph)
max_degree2 = np.max(theme_graph2)
```

```python
norm1 = theme_graph / max_degree1
norm1_sum = norm1.sum(axis = 0)

norm2 = theme_graph2 / max_degree2
norm2_sum = norm2.sum(axis = 1)

ratio = norm2_sum / norm1_sum

up_list = []
for them, i in zip(lookup_gkg_themes, ratio):
    if not math.isnan(i): #nan   
        up_list.append((i, them)) 
        
up_list.sort(reverse = True)

n = 0
for i in up_list:
    if i[0] > 100000000:
        continue
    else:
        if n < 30:
            print(i)
            n +=1
```
